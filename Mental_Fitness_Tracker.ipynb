```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in data.columns:
    if data[i].dtype == 'object':
        data[i]=l.fit_transform(data[i])
```

```python
A = data.drop('Anxiety',axis=1)
b = data['Anxiety']

from sklearn.model_selection import train_test_split
Atrain, Atest, btrain, btest = train_test_split(A, b, test_size=0.2, random_state=2)
```

# RANDOM FOREST REGRESSOR

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import numpy as np
rf = RandomForestRegressor()
rf.fit(Atrain, btrain)
btrain_pred = rf.predict(Atrain)
mse_rf = mean_squared_error(btrain, btrain_pred)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(btrain, btrain_pred)

print(mse_rf)
print(rmse_rf)
print(r2_rf)
```

# Testing

```python
# model evaluation for testing set
btest_pred = rf.predict(Atest)
mse = mean_squared_error(btest, btest_pred)
rmse = (np.sqrt(mean_squared_error(btest, btest_pred)))
r2 = r2_score(btest, btest_pred)
```
